```python
def pluralize_gift(gifts):
    pluralized_gifts = []

    for gift in gifts:

        if gift[-1] == 's':
            pluralized_gifts.append(gift + 'es')
        elif gift[-1] == 'y':
            pluralized_gifts.append(gift[:-1] + 'ies')
        else:
            pluralized_gifts.append(gift + 's')
    return pluralized_gifts

pluralize_gift(xmas['Gift.Item'])
```

```python
def make_phrase(Day, Day_in_Words, Gift_Item, Verb, Adjective, Location):
    Verb = Verb or ""
    Adjective = Adjective or ""

    if Day > 1:
        Gift_Item = pluralize_gift([Gift_Item])[0]
    vowel = Gift_Item[0].lower() in ['a', 'e', 'i', 'o', 'u']

    if Day == 1:
        if vowel:
            day_phrase = "an"
        else:
            day_phrase = "a"
        phrase = f"{Day_in_Words} {Adjective} {Gift_Item} {Verb}"
    else:
        phrase = f"{Day_in_Words} {Gift_Item} {Verb}"

    return phrase
print(phrase)
```

```python
twelvedd =make_phrase(12, 'twelfth', 'drummer', 'drumming', 'Nan', 'NaN')
twelvedd
```

```python
xmas['Full.Phrase'] = xmas.apply(lambda row: make_phrase(row['Day'], row['Day.in.Words'], row['Gift.Item'], row['Verb'], row['Adjective'], row['Location']), axis=1)
```

```python
def sing_day(dataset, num, phrase_col):

    # Step 1: Setup the intro line
    words = ["first", "second", "third", "fourth", "fifth", "sixth", "seventh", "eighth", "ninth", "tenth", "eleventh", "twelfth"]
    word = words[num - 1]
    intro = f"On the {word} day of Christmas, my true love sent to me: "

    gifts = ""
    for i in range(num, 0, -1):
        gift_phrase = dataset.loc[i - 1, phrase_col]
        if i == 1:
            gifts += f"{gift_phrase} and a partridge in a pear tree."
        else:
            gifts += f"{gift_phrase}, "
    return intro + gifts
```

```python
sing_day(xmas, 12, "Full.Phrase")
```
